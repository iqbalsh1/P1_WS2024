## pair0060: 
UNdata: life expectancy at birth for different countries, male, 2000-2005

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0060.txt', delimiter='\t', header=None, names=['life expectancy', 'latitude'])  
```

```python
# Display few rows of the dataset
print(df.head())    
```

```python
# Scatter Plot
plt.scatter(df['life expectancy'], df['latitude'], color = 'purple')
plt.xlabel('life expectancy at birth for different countries')
plt.ylabel('latitude of country capital')
plt.title('Scatter plot of life expectancy vs latitude')
plt.show()
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for (X -> Y)
sns.histplot(df['life expectancy'], kde=True, ax=axis[0], color = 'green')
axis[0].set_xlabel('life expectancy')
axis[0].set_ylabel('Frequency')
axis[0].set_title('life expectancy distribution with KDE')

# Plot (Y -> X)
sns.histplot(df['latitude'], kde=True, ax=axis[1], color = 'orange')
axis[1].set_xlabel('latitude')
axis[1].set_ylabel('Frequency')
axis[1].set_title('latitude distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['life expectancy']))
print(f"Dip-Test (X -> Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['latitude']))
print(f"Dip-Test (Y -> X): {dip_test}, p-value: {p_val}")
```

### Findings:
- The ground truth relationship is Y → X
- The dip test results suggest that the distribution of X → Y and Y → X both direction is likely **unimodal**, supported by a high p-value.
