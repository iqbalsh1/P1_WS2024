```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest  
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0035.txt', delimiter=' ', header=None, names=['drinks', 'sgpt'])   
```

```python
# Display few rows of the dataset
print(df.head())
```

```python
# Scatter Plot
plt.scatter(df['drinks'], df['sgpt'])
plt.xlabel('drinks (half-pint per day)')
plt.ylabel('sgpt (alanine aminotransferase)')
plt.title('Scatter plot of drinks vs sgpt')
plt.show() 
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for drinks -> sgpt (X -> Y)
sns.histplot(df['drinks'], kde=True, ax=axis[0])
axis[0].set_xlabel('drinks')
axis[0].set_ylabel('sgpt')
axis[0].set_title('drinks distribution with KDE')

# Plot for sgpt -> drinks (Y -> X)
sns.histplot(df['sgpt'], kde=True, ax=axis[1])
axis[1].set_xlabel('sgpt')
axis[1].set_ylabel('drinks')
axis[1].set_title('sgpt distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['drinks']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['sgpt']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results suggest that the distribution of X → Y is likely **multimodal**, supported by a low p-value 0.
- While the distribution of Y → X direction is likely  **unimodal**, though there is still some uncertainty as this p-value (0.11) doesn't show strong evidence against unimodality.
