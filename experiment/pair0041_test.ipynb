```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0041.txt', delimiter=' ', header=None, names=['age', 'Plasma glucose concentration'])    
```

```python
# Display few rows of the dataset
print(df.head()) 
```

```python
# Scatter Plot
plt.scatter(df['age'], df['Plasma glucose concentration'])
plt.xlabel('age')
plt.ylabel('Plasma glucose concentration')
plt.title('Scatter plot of age vs Plasma glucose concentration')
plt.show()
```

```python

# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for age -> Plasma glucose concentration (X -> Y)
sns.histplot(df['age'], kde=True, ax=axis[0])
axis[0].set_xlabel('age')
axis[0].set_ylabel('Plasma glucose concentration')
axis[0].set_title('age distribution with KDE')

# Plot for Plasma glucose concentration -> age (Y -> X)
sns.histplot(df['Plasma glucose concentration'], kde=True, ax=axis[1])
axis[1].set_xlabel('Plasma glucose concentration')
axis[1].set_ylabel('age')
axis[1].set_title('Plasma glucose concentration distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['age']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['Plasma glucose concentration']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results suggest that the distribution of X → Y is likely **multimodal**, supported by a low p-value 0.
- While the distribution of Y → X direction is **unimodal**, supported by a high p-value 0.67.
