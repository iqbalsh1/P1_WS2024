## pair0004: 
#### Weather Data from 349 DWD (Deutscher Wetterdienst) Stations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest   
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0004.txt', delimiter=' ', header=None, names=['altitude', 'sunshine'])
```

```python
# Display few rows of the dataset
print(df.head())   
```

```python
# Scatter Plot
plt.scatter(df['altitude'], df['sunshine'])
plt.xlabel('altitude')
plt.ylabel('sunshine')
plt.title('Scatter plot (averaged over 1961-1990)')
plt.show()  
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for altitude -> sunshine (X -> Y)
sns.histplot(df['altitude'], kde=True, ax=axis[0])
axis[0].set_xlabel('altitude')
axis[0].set_ylabel('sunshine')
axis[0].set_title('altitude distribution with KDE')

# Plot for altitude -> sunshine (Y -> X)
sns.histplot(df['sunshine'], kde=True, ax=axis[1])
axis[1].set_xlabel('sunshine')
axis[1].set_ylabel('altitude')
axis[1].set_title('sunshine distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()   
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['altitude']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['sunshine']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")    
```

### Findings:
- The ground truth relationship is 𝑋 → 𝑌.
- The dip test results show no multimodality in both direction (𝑋 → 𝑌 or 𝑌 → 𝑋 )
- since both p-values are high, exhibiting that the distributions are likely **unimodal**.
