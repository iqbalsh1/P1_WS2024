## pair0010: 
Abalone data: Abalones are marine snails

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
    
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0010.txt', delimiter=' ', header=None, names=['rings', 'shucked weigth'])   
```

```python
# Display few rows of the dataset
print(df.head())
```

```python
# Scatter Plot
plt.scatter(df['rings'], df['shucked weigth'])
plt.xlabel('rings')
plt.ylabel('shucked weigth grams')
plt.title('Scatter plot of rings vs shucked weigth')
plt.show()
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for rings -> shucked weigth (X -> Y)
sns.histplot(df['rings'], kde=True, ax=axis[0])
axis[0].set_xlabel('rings')
axis[0].set_ylabel('shucked weigth grams')
axis[0].set_title('rings distribution with KDE')

# Plot for rings -> shucked weigth (Y -> X)
sns.histplot(df['shucked weigth'], kde=True, ax=axis[1])
axis[1].set_xlabel('shucked weigth grams')
axis[1].set_ylabel('rings')
axis[1].set_title('shucked weigth distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()   
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['rings']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['shucked weigth']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")   
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results indicate a multimodal distribution for X → Y, with a p-value of zero, confirming **multimodality.**
- In contrast, the distribution for Y → X is likely **unimodal**, supported by a high p-value of 0.90.
