```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest  
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0040.txt', delimiter=' ', header=None, names=['age', 'diastolic blood pressure'])  
```

```python
# Display few rows of the dataset
print(df.head())
    
```

```python
# Scatter Plot
plt.scatter(df['age'], df['diastolic blood pressure'])
plt.xlabel('age')
plt.ylabel('diastolic blood pressure (mm Hg)')
plt.title('Scatter plot of age vs diastolic blood pressure')
plt.show()
    
```

```python

# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for age -> diastolic blood pressure (X -> Y)
sns.histplot(df['age'], kde=True, ax=axis[0])
axis[0].set_xlabel('age')
axis[0].set_ylabel('diastolic blood pressure')
axis[0].set_title('age distribution with KDE')

# Plot for diastolic blood pressure -> age (Y -> X)
sns.histplot(df['diastolic blood pressure'], kde=True, ax=axis[1])
axis[1].set_xlabel('diastolic blood pressure')
axis[1].set_ylabel('age')
axis[1].set_title('diastolic blood pressure distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['age']))
print(f"Dip-Test (X -> Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['diastolic blood pressure']))
print(f"Dip-Test (Y -> X): {dip_test}, p-value: {p_val}")
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results suggest that the distribution of X → Y is likely **multimodal**, supported by a low p-value 0.
- Similarly the distribution of Y → X direction is also **multimodal**, supported by a ver low p-value 2.09e-06, providing strong evidence against unimodality.
