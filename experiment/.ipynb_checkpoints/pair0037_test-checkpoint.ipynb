```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
    
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0037.txt', delimiter=' ', header=None, names=['drinks', 'gammagt'])   
```

```python
# Display few rows of the dataset
print(df.head())   
```

```python
# Scatter Plot
plt.scatter(df['drinks'], df['gammagt'])
plt.xlabel('drinks (half pint per day)')
plt.ylabel('gammagt (gamma-glutamyl transpeptdase)')
plt.title('Scatter plot of drinks vs gammagt')
plt.show()
    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for drinks -> gammagt (X -> Y)
sns.histplot(df['drinks'], kde=True, ax=axis[0])
axis[0].set_xlabel('drinks')
axis[0].set_ylabel('gammagt')
axis[0].set_title('drinks distribution with KDE')

# Plot for gammagt -> drinks (Y -> X)
sns.histplot(df['gammagt'], kde=True, ax=axis[1])
axis[1].set_xlabel('gammagt')
axis[1].set_ylabel('drinks')
axis[1].set_title('gammagt distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['drinks']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['gammagt']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results suggest that the distribution of X → Y is likely **multimodal**, supported by a low p-value 0.
- While the distribution of Y → X direction is **unimodal**, supported by a high p-value 0.44.
