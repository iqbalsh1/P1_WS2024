## pair0009: 
Abalone data: Abalones are marine snails


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
    
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0009.txt', delimiter=' ', header=None, names=['rings', 'whole weight'])
```

```python
# Display few rows of the dataset
print(df.head()) 
```

```python
# Scatter Plot
plt.scatter(df['rings'], df['whole weight'])
plt.xlabel('rings')
plt.ylabel('whole weight grams')
plt.title('Scatter plot of rings vs whole weight')
plt.show()   
```

"Rings" represent the no. of growth rings in the abalone's shell. It is used to estimate the age of the abalone i.e. by adding 1.5 to the number of rings. 
- Age (in years) = Rings + 1.5

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for rings -> whole weight (X -> Y)
sns.histplot(df['rings'], kde=True, ax=axis[0])
axis[0].set_xlabel('rings')
axis[0].set_ylabel('whole weight grams')
axis[0].set_title('rings distribution with KDE')

# Plot for rings -> whole weight (Y -> X)
sns.histplot(df['whole weight'], kde=True, ax=axis[1])
axis[1].set_xlabel('whole weight grams')
axis[1].set_ylabel('rings')
axis[1].set_title('whole weight distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['rings']))
print(f"Dip-Test (X  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['whole weight']))
print(f"Dip-Test (Y  X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is X  Y
- The dip test results indicate a multimodal distribution for X  Y, with a p-value of zero, confirming **multimodality.**
- In contrast, the distribution for Y  X is likely **unimodal**, supported by a higher p-value of 0.50..
