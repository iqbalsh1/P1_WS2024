## pair0047: 
Traffic: cars per 24h at different counting stations in Oberschwaben, Germany


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0047.txt', delimiter='\t', header=None, names=['number of cars', 'category'])
```

```python
# Display few rows of the dataset
print(df.head())
    
```

```python
# Scatter Plot
plt.scatter(df['category'], df['number of cars'])
plt.xlabel('number of cars')
plt.ylabel('category')
plt.title('Scatter plot of number of cars vs category (holidays/working days)')
plt.show()  
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for number of cars -> category (X -> Y)
sns.histplot(df['number of cars'], kde=True, ax=axis[0])
axis[0].set_xlabel('number of cars')
axis[0].set_ylabel('category')
axis[0].set_title('number of cars distribution with KDE')

# Plot for category -> number of cars (Y -> X)
sns.histplot(df['category'], kde=True, ax=axis[1])
axis[1].set_xlabel('category')
axis[1].set_ylabel('number of cars')
axis[1].set_title('category distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['number of cars']))
print(f"Dip-Test (X  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['category']))
print(f"Dip-Test (Y  X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is Y  X
- The dip test results suggest that the distribution of X  Y and Y  X are both likely **multimodal**, supported by a low p-value 0.
