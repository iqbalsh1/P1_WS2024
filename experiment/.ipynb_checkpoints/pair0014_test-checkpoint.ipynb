## pair0014: 
Auto-Mpg Data: city-cycle fuel consumption data in miles per gallon

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest  
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0014.txt', delimiter=' ', header=None, names=['mpg', 'horsepower'])    
```

```python
# Display few rows of the dataset
print(df.head())    
```

```python
# Scatter Plot
plt.scatter(df['mpg'], df['horsepower'])
plt.xlabel('mpg')
plt.ylabel('horsepower')
plt.title('Scatter plot of mpg vs horsepower')
plt.show()    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for mpg -> horsepower (X -> Y)
sns.histplot(df['mpg'], kde=True, ax=axis[0])
axis[0].set_xlabel('mpg')
axis[0].set_ylabel('horsepower')
axis[0].set_title('mpg distribution with KDE')

# Plot for mpg -> horsepower (Y -> X)
sns.histplot(df['horsepower'], kde=True, ax=axis[1])
axis[1].set_xlabel('horsepower')
axis[1].set_ylabel('mpg')
axis[1].set_title('horsepower distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()   
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['mpg']))
print(f"Dip-Test (X  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['horsepower']))
print(f"Dip-Test (Y  X): {dip_test}, p-value: {p_val}")   
```

### Findings:
- The ground truth relationship is X  Y
- The dip test results indicate that the distribution for X  Y is likely **multimodal**, with a p-value of 0.0303.
- In contrast, the distribution for Y  X is likely **unimodal**, supported by a p-value of 0.1139.
