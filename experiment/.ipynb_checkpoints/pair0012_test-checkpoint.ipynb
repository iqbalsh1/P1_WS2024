## pair0012: 

Census Income (KDD) dataset: from the 1994 and 1995
current population surveys conducted by the U.S. Census Bureau.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest 
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0012.txt', delimiter=' ', header=None, names=['age', 'wage per hour'])  
```

```python
# Display few rows of the dataset
print(df.head())   
```

```python
# Scatter Plot
plt.scatter(df['age'], df['wage per hour'])
plt.xlabel('age')
plt.ylabel('wage per hour')
plt.title('Scatter plot of age vs wage per hour')
plt.show()    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for age -> wage per hour (X -> Y)
sns.histplot(df['age'], kde=True, ax=axis[0])
axis[0].set_xlabel('age')
axis[0].set_ylabel('wage per hour')
axis[0].set_title('age distribution with KDE')

# Plot for age -> wage per hour (Y -> X)
sns.histplot(df['wage per hour'], kde=True, ax=axis[1])
axis[1].set_xlabel('wage per hour')
axis[1].set_ylabel('wage')
axis[1].set_title('wage per hour distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()   
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['age']))
print(f"Dip-Test (X  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['wage per hour']))
print(f"Dip-Test (Y  X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is X  Y
- The dip test results indicate the distribution in both directions    and    is **multimodal**, based on the low p-value = 0 .
