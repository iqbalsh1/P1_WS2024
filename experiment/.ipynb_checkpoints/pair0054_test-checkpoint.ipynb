## pair0054: 
Auto-Mpg Data


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
```

```python
# Define column 
columns = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']

# Load the dataset
df = pd.read_csv('datasets/pair0054.txt', sep=r'\s+', header=None, names=columns)

# Separate columns into x and y 
x = df[['displacement', 'horsepower', 'weight']]
y = df[['mpg', 'acceleration']]
```

```python
# Display few rows of the dataset
print(df.head())
```

```python
plt.figure(figsize=(10, 6))

# Define clrs for each scatter plot
clrs = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
labels = ['Dis vs Mpg', 'Hp vs. Mpg', 'W vs Mpg','Dis vs Acc', 'Hp vs Acc', 'W vs Acc']

plt.scatter(x['displacement'], y['mpg'], color=clrs[0], label= labels[0])
plt.scatter(x['horsepower'], y['mpg'], color=clrs[1], label= labels[1])
plt.scatter(x['weight'], y['mpg'], color=clrs[2], label= labels[2])
plt.scatter(x['displacement'], y['acceleration'], color=clrs[3], label= labels[3])
plt.scatter(x['horsepower'], y['acceleration'], color=clrs[4], label=labels[4])
plt.scatter(x['weight'], y['acceleration'], color=clrs[5], label= labels[5])

# Add labels and title
plt.xlabel('Values of X (Displacement, Horsepower, Weight)')
plt.ylabel('Values of Y (Mpg, Acceleration)')
plt.title('Scatter plot of Auto-Mpg Data')
plt.legend()
plt.show()
```

```python
# Combine all values of x and y 
combinedX = pd.concat([x['displacement'], x['horsepower'], x['weight']])
combinedY = pd.concat([y['mpg'], y['acceleration']])
```

```python
fig, axis = plt.subplots(1, 2, figsize=(16, 6))

# Plot for x (displacement, horsepower, weight) -> y (mpg, acceleration)
sns.histplot(combinedX, kde=True, ax=axis[0], color='green')
axis[0].set_xlabel('MPG')
axis[0].set_ylabel('Frequency')
axis[0].set_title('MPG Distribution with KDE')

# Plot for y (mpg, acceleration) -> x (displacement, horsepower, weight)
sns.histplot(combinedY, kde=True, ax=axis[1], color='yellow')
axis[1].set_xlabel('Acceleration')
axis[1].set_ylabel('Frequency')
axis[1].set_title('Acceleration Distribution with KDE')

plt.tight_layout()
plt.show()
    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(combinedX))
print(f"Dip-Test (X  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(combinedY))
print(f"Dip-Test (Y  X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is X  Y
- The dip test results suggest that the distribution of X  Y and Y  X direction is likely **unimodal**, supported by a low p-value 0.
