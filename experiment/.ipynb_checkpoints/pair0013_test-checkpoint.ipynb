## pair0013: 
Auto-Mpg Data: city-cycle fuel consumption data in miles per gallon


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
    
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0013.txt', delimiter=' ', header=None, names=['displacement', 'mpg'])   
```

```python
# Display few rows of the dataset
print(df.head())
```

```python
# Scatter Plot
plt.scatter(df['displacement'], df['mpg'])
plt.xlabel('displacement')
plt.ylabel('mpg')
plt.title('Scatter plot of displacement vs mpg')
plt.show()
    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for displacement -> mpg (X -> Y)
sns.histplot(df['displacement'], kde=True, ax=axis[0])
axis[0].set_xlabel('displacement')
axis[0].set_ylabel('mpg')
axis[0].set_title('displacement distribution with KDE')

# Plot for displacement -> mpg (Y -> X)
sns.histplot(df['mpg'], kde=True, ax=axis[1])
axis[1].set_xlabel('mpg')
axis[1].set_ylabel('displacement')
axis[1].set_title('mpg distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['displacement']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['mpg']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results indicate that the distribution of X → Y is **multimodal**, supported by a p-value of 0.0, confirming multimodality.
- In contrast, the distribution for Y → X is likely **unimodal**, supported by a high p-value of 0.11
