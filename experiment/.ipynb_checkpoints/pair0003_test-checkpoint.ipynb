## pair0003: 
#### Weather Data from 349 DWD (Deutscher Wetterdienst) Stations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
    
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0003.txt', delimiter=' ', header=None, names=['longitude', 'temperature'])   
```

```python
# Display few rows of the dataset
print(df.head())    
```

```python
# Scatter Plot
plt.scatter(df['longitude'], df['temperature'])
plt.xlabel('longitude')
plt.ylabel('temperature')
plt.title('Scatter plot (averaged over 1961-1990)')
plt.show()    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for longitude -> temperature (X -> Y)
sns.histplot(df['longitude'], kde=True, ax=axis[0])
axis[0].set_xlabel('longitude')
axis[0].set_ylabel('temperature')
axis[0].set_title('longitude distribution with KDE')

# Plot for longitude -> temperature (Y -> X)
sns.histplot(df['temperature'], kde=True, ax=axis[1])
axis[1].set_xlabel('temperature')
axis[1].set_ylabel('longitude')
axis[1].set_title('temperature distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['longitude']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['temperature']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is 𝑋 → 𝑌.
- The dip test results show no multimodality in both direction (𝑋 → 𝑌 or 𝑌 → 𝑋 )
- since both p-values are high, exhibiting that the distributions are likely **unimodal**.
