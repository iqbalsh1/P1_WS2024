```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest   
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0036.txt', delimiter=' ', header=None, names=['drinks', 'sgot'])   
```

```python
# Display few rows of the dataset
print(df.head())  
```

```python
# Scatter Plot
plt.scatter(df['drinks'], df['sgot'])
plt.xlabel('drinks (half-pint per day)')
plt.ylabel('sgot (aspartate aminotransferase)')
plt.title('Scatter plot of drinks vs sgot')
plt.show()
    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for drinks -> sgot (X -> Y)
sns.histplot(df['drinks'], kde=True, ax=axis[0])
axis[0].set_xlabel('drinks')
axis[0].set_ylabel('sgot')
axis[0].set_title('drinks distribution with KDE')

# Plot for sgot -> drinks (Y -> X)
sns.histplot(df['sgot'], kde=True, ax=axis[1])
axis[1].set_xlabel('sgot')
axis[1].set_ylabel('drinks')
axis[1].set_title('sgot distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['drinks']))
print(f"Dip-Test (X  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['sgot']))
print(f"Dip-Test (Y  X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is X  Y
- The dip test results suggest that the distribution of X  Y is likely **multimodal**, supported by a low p-value 0.
- While the distribution of Y  X direction also suggest **multimodal**, supported by a low p-value 0.007.
