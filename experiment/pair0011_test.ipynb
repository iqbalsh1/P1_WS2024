## pair0011: 
- Abalone data: Abalones are marine snails
- Viscera Weight = gut weight (after bleeding)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest
    
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0011.txt', delimiter=' ', header=None, names=['rings', 'viscera weight'])   
```

```python
# Display few rows of the dataset
print(df.head())   
```

```python
# Scatter Plot
plt.scatter(df['rings'], df['viscera weight'])
plt.xlabel('rings')
plt.ylabel('viscera weight grams')
plt.title('Scatter plot of rings vs viscera weight')
plt.show()    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for rings -> viscera weight (X -> Y)
sns.histplot(df['rings'], kde=True, ax=axis[0])
axis[0].set_xlabel('rings')
axis[0].set_ylabel('viscera weight (grams)')
axis[0].set_title('rings distribution with KDE')

# Plot for rings -> viscera weight (Y -> X)
sns.histplot(df['viscera weight'], kde=True, ax=axis[1])
axis[1].set_xlabel('viscera weight (grams)')
axis[1].set_ylabel('rings')
axis[1].set_title('viscera weight distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['rings']))
print(f"Dip-Test (X  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['viscera weight']))
print(f"Dip-Test (Y  X): {dip_test}, p-value: {p_val}")
```

### Findings:
- The ground truth relationship is X  Y
- The dip test results indicate a **multimoda**l distribution for X  Y, supported by a p-value of 0.0, confirming multimodality.
- In contrast, the distribution for Y  X is likely **unimodal**, supported by a high p-value of 0.8446.
