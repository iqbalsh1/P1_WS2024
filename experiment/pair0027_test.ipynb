Concrete Compressive Strength 


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest    
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0027.txt', delimiter=' ', header=None, names=['fly ash', 'compressive strength'])    
```

```python
# Display few rows of the dataset
print(df.head())
```

```python
# Scatter Plot
plt.scatter(df['fly ash'], df['compressive strength'])
plt.xlabel('fly ash (kg in m3 mixture)')
plt.ylabel('compressive strength (MPa)')
plt.title('Scatter plot of fly ash vs compressive strength')
plt.show()    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for fly ash -> compressive strength (X -> Y)
sns.histplot(df['fly ash'], kde=True, ax=axis[0])
axis[0].set_xlabel('fly ash')
axis[0].set_ylabel('compressive strength')
axis[0].set_title('fly ash distribution with KDE')

# Plot for compressive strength -> fly ash (Y -> X)
sns.histplot(df['compressive strength'], kde=True, ax=axis[1])
axis[1].set_xlabel('compressive strength')
axis[1].set_ylabel('fly ash')
axis[1].set_title('compressive strength distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()   
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['fly ash']))
print(f"Dip-Test (X →  Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['compressive strength']))
print(f"Dip-Test (Y →  X): {dip_test}, p-value: {p_val}")  
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results suggest that the distribution of X → Y is likely **multimodal**, supported by a low p-value 0.
- While the distribution of Y → X direction is **unimodal**, supported by a high p-value 0.56.
