## pair0050: 
Daily mean values of ozone and temperature of year 2009 in Chaumont, Switzerland

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest  
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0050.txt', sep=r'\s+', header=None, names=['ozone', 'temperature'])   
```

```python
# Display few rows of the dataset
print(df.head())
```

```python
# Scatter Plot
plt.scatter(df['ozone'], df['temperature'], color = 'blue')
plt.xlabel('ozone (microgram / cubic meter)')
plt.ylabel('temperature (degree celsius)')
plt.title('Scatter plot of ozone vs temperature')
plt.show()
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for ozone -> temperature (X -> Y)
sns.histplot(df['ozone'], kde=True, ax=axis[0], color = 'green')
axis[0].set_xlabel('ozone')
axis[0].set_ylabel('temperature')
axis[0].set_title('ozone distribution with KDE')

# Plot for temperature -> ozone (Y -> X)
sns.histplot(df['temperature'], kde=True, ax=axis[1], color = 'orange')
axis[1].set_xlabel('temperature')
axis[1].set_ylabel('ozone')
axis[1].set_title('temperature distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()   
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['ozone']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['temperature']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")
```

### Findings:
- The ground truth relationship is Y → X
- The dip test results suggest that the distribution of X → Y is likely **unimodal**, supported by a high p-value 0.990.
- Similarly, the distribution of Y → X direction is also **unimodal**, supported by a high p-value 0.199.
