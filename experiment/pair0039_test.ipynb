```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from diptest import diptest  
```

```python
# Load dataset
df = pd.read_csv('datasets/pair0039.txt', delimiter=' ', header=None, names=['age', '2-Hour serum insulin'])   
```

```python
# Display few rows of the dataset
print(df.head())    
```

```python
# Scatter Plot
plt.scatter(df['age'], df['2-Hour serum insulin'])
plt.xlabel('age')
plt.ylabel('2-Hour serum insulin (mu U/ml)')
plt.title('Scatter plot of age vs 2-Hour serum insulin')
plt.show()
    
```

```python
# Plot the distribution of the data in both directions (x and y)
fig, axis = plt.subplots(1, 2, figsize=(16,6))

# Plot for age -> 2-Hour serum insulin (X -> Y)
sns.histplot(df['age'], kde=True, ax=axis[0])
axis[0].set_xlabel('age')
axis[0].set_ylabel('2-Hour serum insulin')
axis[0].set_title('age distribution with KDE')

# Plot for 2-Hour serum insulin -> age (Y -> X)
sns.histplot(df['2-Hour serum insulin'], kde=True, ax=axis[1])
axis[1].set_xlabel('2-Hour serum insulin')
axis[1].set_ylabel('age')
axis[1].set_title('2-Hour serum insulin distribution with KDE')

# Adjust layout to set the title and labels
plt.tight_layout()
plt.show()
    
```

```python
# Dip-Test for x and y
dip_test, p_val = diptest(np.array(df['age']))
print(f"Dip-Test (X → Y): {dip_test}, p-value: {p_val}")   
dip_test, p_val = diptest(np.array(df['2-Hour serum insulin']))
print(f"Dip-Test (Y → X): {dip_test}, p-value: {p_val}")
    
```

### Findings:
- The ground truth relationship is X → Y
- The dip test results suggest that the distribution of X → Y is likely **multimodal**, supported by a very low p-value 9.94e-06.
- While the distribution of Y → X direction is **unimodal**, supported by a high p-value 0.93.
